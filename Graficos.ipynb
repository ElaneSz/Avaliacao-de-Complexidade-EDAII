```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data_avl = pd.read_excel("/content/sample_data/avl.xlsx")
print(data_avl)

caso_medio_avl = data_avl.iloc[1:20,[2, 3]]
pior_caso_medio_avl = data_avl.iloc[1:20,[0, 1]]

caso_medio_avl = np.array(caso_medio_avl) 
pior_caso_medio_avl = np.array(pior_caso_medio_avl)
```

```python
data_b = pd.read_excel("/content/sample_data/arvore_b.xlsx")
print(data_b)

caso_medio_b = data_b.iloc[1:20,[2, 3]]
pior_caso_medio_b = data_b.iloc[1:20,[0, 1]]

caso_medio_b = np.array(caso_medio_b) 
pior_caso_medio_b = np.array(pior_caso_medio_b)
```

```python
data_rubro = pd.read_excel("/content/sample_data/rubro_negra.xlsx")
print(data_rubro)

caso_medio_rubro = data_rubro.iloc[1:20,[2, 3]]
pior_caso_medio_rubro = data_rubro.iloc[1:20,[0, 1]]

caso_medio_rubro = np.array(caso_medio_rubro) 
pior_caso_medio_rubro = np.array(pior_caso_medio_rubro)
```

```python
df_avl = pd.DataFrame(caso_medio_avl, columns = ['numero_dados', 'iteracoes'])
df_arvore_b = pd.DataFrame(caso_medio_b, columns = ['numero_dados', 'iteracoes'])
df_rubro = pd.DataFrame(caso_medio_rubro, columns = ['numero_dados', 'iteracoes'])
```

```python
g1 = sns.lineplot(data=df_avl, x ='numero_dados', y ='iteracoes', color='red', label='AVL').set(title='Caso Médio')
g1 = sns.lineplot(data=df_arvore_b, x ='numero_dados', y ='iteracoes', color='orange', label='Árvore-B').set(title='Caso Médio')
g1 = sns.lineplot(data=df_rubro, x ='numero_dados', y ='iteracoes', color='yellow', label='Rubro-Negra').set(title='Caso Médio')
```
